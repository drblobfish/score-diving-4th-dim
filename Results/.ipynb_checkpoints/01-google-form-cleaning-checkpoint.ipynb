```python
def translate(x):
  if type(x) is str and (not x.isnumeric()):
    return ";".join(trad[answ] for answ in x.split(";") )
  else :
    return x
```

```python
startFormFraTranslated = startFormFra.drop("Timestamp",axis=1).applymap(translate)
```

now we can concatenate them

```python
startForm = pd.concat([startFormFraTranslated,startFormEng],axis=0, ignore_index=True)
```
